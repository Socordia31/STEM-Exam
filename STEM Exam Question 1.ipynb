```python
# Question 1
colour_im = np.array(Image.open('images/robot-enstein.jpg'))

# Your code here
for i in colour_im:  # get pixels in one row
    for j in i:  # get each pixel
        j[0] = int(j[0]*0.8)  # reduce the red channel by 0.2
        # increase the blue channel by 0.15 but not exceed the upper bound
        j[2] = min(int(j[2]*1.15), 255)

Image.fromarray(colour_im)
```
